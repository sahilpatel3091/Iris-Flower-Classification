```python
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv(r"D:\.1--MCA--MSU--\MCA_SEM_3\Machine Learning\Data_Set\IRIS_flower_logistic_reg.csv")
```

```python
df.head(10)
```

```python
#sepal_lenght, sepal_width, petal_width, petal_length are independent feature while species are dependent variable
#sepal = lower part of flower
#petal = upper part of flower
df['species'].value_counts()
```

```python
flower_map = {'Iris-setosa' : 0, 'Iris-versicolor' : 1, 'Iris-virginica' : 2 }
df["species"] = df["species"].map(flower_map)
```

```python
df.head()
```

```python
custom_palette = ['#1f77b4', '#ff7f0e', '#2ca02c']
sns.pairplot(df, hue = 'species', palette =custom_palette )
```

```python
X = df[['sepal_length',	'sepal_width', 'petal_length',	'petal_width']].values
y = df[['species']].values
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X,y)
```

```python
model.score(X,y)
```

```python
exp = y
pre = model.predict(X)
pre
```

```python
from sklearn import metrics
```

```python
print(metrics.classification_report(exp,pre))
```

```python
print(metrics.confusion_matrix(exp,pre))
```
